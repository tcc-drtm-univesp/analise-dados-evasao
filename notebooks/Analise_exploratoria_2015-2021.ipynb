```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
dtypes = {
    'Ano de Ingresso': int,      # String
    'Ano de Referência': int,     # Inteiro
}
df = pd.read_excel("indicadores_trajetoria_educacao_superior_2015_2021.xlsx", dtype=dtypes)
df
```

```python
def DesistenciaMediaGrandeAreaporAno(dfa, ano):
    """
    Função para calcular a média de desistência por Grande Área do CINE Brasil.
    Os parâmetros são o dataframe dfa e o ano desejado em inteiros.
    Por exemplo, para o primeiro ano de curso chamar DesistenciaMediaGrandeAreaporAno(df, 1)
    """
    #Filtra o df pelo ano
    dfa = dfa[dfa['Ano de Ingresso']==dfa['Ano de Referência'] - (ano-1)]
    
    #Imprime informações do df filtrado
    print(dfa.info())
    
    #Agrupa a desistência por Grande Área e imprime
    media_por_grande_area_ano = dfa.groupby('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')['Taxa de Desistência Anual - TADA'].mean()
    print(media_por_grande_area_ano)
    
    #Exibe gráfico
    media_por_grande_area_ano.plot(kind='bar')
    plt.xlabel('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')
    plt.ylabel('Taxa de Desistência Anual - TDAN')
    title = 'TDAN Média por Grande Área no ano ' + str(ano)
    plt.title(title)
```

```python
def DesistenciaMediaGrandeAreaConsideraPandemia(dfa, pandemia):
    """
    Função para calcular a média de desistência por Grande Área do CINE Brasil.
    Os parâmetros são o dataframe dfa e se deseja considerar anos com ou sem pandemia
    Por exemplo, para anos "sem" pandemia chamar DesistenciaMediaGrandeAreaConsideraPandemia(dfa, "não")
    Por exemplo, para anos "com" pandemia chamar DesistenciaMediaGrandeAreaConsideraPandemia(dfa, "sim")
    """
    
    if pandemia == "não":
        #Filtra apenas anos sem pandemia
        dfa_sem_pandemia = dfa[(dfa["Ano de Referência"] != 2020) & (dfa["Ano de Referência"] != 2021) ]
        print(dfa_sem_pandemia.info())
        #Agrupa por Grande Área CINE
        media_por_grande_area_sem_pandemia = dfa_sem_pandemia.groupby('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')['Taxa de Desistência Anual - TADA'].mean()
        
        #Imprime média por Grande Área
        print(media_por_grande_area_sem_pandemia)
        
        #Exibe gráfico
        media_por_grande_area_sem_pandemia.plot(kind='bar')
        plt.xlabel('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')
        plt.ylabel('Taxa de Desistência Anual - TDAN')
        plt.title('TDAN Média por Grande Área em anos sem pandemia')
        
    if pandemia == "sim":
        #Filtra apenas anos com pandemia
        dfa_com_pandemia = dfa[(dfa["Ano de Referência"] == 2020) | (dfa["Ano de Referência"] == 2021)]
        print(dfa_com_pandemia.info())
        
        #Agrupa por Grande Área CINE
        media_por_grande_area_com_pandemia = dfa_com_pandemia.groupby('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')['Taxa de Desistência Anual - TADA'].mean()
        
        #Exibe gráfico
        print(media_por_grande_area_com_pandemia)
        media_por_grande_area_com_pandemia.plot(kind='bar')
        plt.xlabel('Nome da Grande Área do Curso segundo a classificação CINE BRASIL')
        plt.ylabel('Taxa de Desistência Anual - TDAN')
        plt.title('TDAN Média por Grande Área em anos com pandemia')
    
```

```python
DesistenciaMediaGrandeAreaConsideraPandemia(df, pandemia = "não")
```

```python
DesistenciaMediaGrandeAreaConsideraPandemia(df, pandemia = "sim")
```
