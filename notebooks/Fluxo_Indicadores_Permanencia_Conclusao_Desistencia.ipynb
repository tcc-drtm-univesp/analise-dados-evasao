```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None)
```

```python
dtypes = {
    'Ano de Ingresso': int,      # String
    'Ano de Referência': int,     # Inteiro
}
df = pd.read_csv("20230809_DadosINEP.csv", dtype=dtypes)
```

```python
df["Ano de Ingresso"].unique()
```

```python
df['TDAN'] = df['Taxa de Desistência Anual - TDAN'].combine_first(df['Taxa de Desistência Anual - TADA'])
```

```python
df['TDA'] = df['Taxa de Desistência Acumulada - TDA'].combine_first(df['Taxa de Desistência Acumulada - TODA'])
```

```python
df
```

```python
df.TDA.info()
```

```python
def IndicadoresporCursoIngresso(dfa, codcurso, ingresso):
    '''
    Função para mostrar gráfico para um determinado curso com os indicadores de Permanência, Conclusão e Desistência
    Os parâmetros são o Dataframe dfa, o Código Do Curso codcurso e o Ano de Ingresso ingresso no formato AAAA
    '''
    #
    dfa = df[(df['Código do Curso de Graduação'] == codcurso) & (dfa['Ano de Ingresso'] == ingresso)]
    dfa = dfa.reset_index()
    nomeCurso = dfa['Nome do Curso de Graduação'][0]
    instituicao = dfa['Nome da Instituição'][0] 
    ingresso = dfa['Ano de Ingresso'][0]
    dfa = dfa[['Ano de Referência','Taxa de Permanência - TAP', 'Taxa de Conclusão Acumulada - TCA', 'TDA']]
    plt.plot(dfa['Ano de Referência'], dfa['Taxa de Permanência - TAP'], label='Taxa de Permanência - TAP')
    plt.plot(dfa['Ano de Referência'], dfa['Taxa de Conclusão Acumulada - TCA'], label='Taxa de Conclusão Acumulada - TCA')
    plt.plot(dfa['Ano de Referência'], dfa['TDA'], label='Taxa de Desistência Acumulada')
    plt.xticks(dfa['Ano de Referência'])
    plt.yticks(range(0, 101, 10))
    plt.ylim(0, 100)
    plt.title(f'Indicadores de Permanência, Conclusão e Desistência do curso {nomeCurso}, {instituicao}, {ingresso}')
    plt.legend()
    plt.show()
```

```python
IndicadoresporCursoIngresso(df, codcurso = 2, ingresso = 2017)
```
