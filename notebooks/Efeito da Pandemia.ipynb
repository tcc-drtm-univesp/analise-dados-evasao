```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
dtypes = {
    'Ano de Ingresso': int,      # String
    'Ano de Referência': int,     # Inteiro
}
df = pd.read_csv("20230809_DadosINEP.csv")
df
```

# Estudo em relação ao efeito da pandemia no 4º ano Geral

```python
#Filtrando apenas pelo dados de 4º ano
df_4ano = df[ df['Ano de Referência'] - df['Ano de Ingresso'] == 3]
df_4ano
```

```python
desistencia_media_por_turma_ingresso = df_4ano.groupby('Ano de Ingresso')['Taxa de Desistência Anual'].mean()
print(desistencia_media_por_turma_ingresso)
    
#Exibe gráfico
desistencia_media_por_turma_ingresso.plot(kind='bar')
plt.xlabel('Ano de Ingresso')
plt.ylabel('Taxa de Desistência Anual - TDAN no 4º ano')
title = 'TDAN Média por Ano no 4º de Ingresso'
plt.title(title)
```

```python
#Calculando a média de desistência no 4º ano para turmas que não tiveram pandemia
desistencia_media_por_turma_ingresso[:-1].mean()
```
